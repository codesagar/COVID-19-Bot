## Preparing questions for webapplication

```python
import pandas as pd
from ast import literal_eval

lang_list = ['English', 'हिंदी', 'ગુજરાતી', 'বাংলা', 'தமிழ்', 'मराठी','తెలుగు', 'ಕನ್ನಡ']
questions = pd.read_csv("final_questions.csv")
print(questions.shape)
options = pd.read_csv("final_options.csv")
print(options.shape)
```

# Sanity checks

```python
option_counts = options.copy()
for col in lang_list:
    option_counts[col] = options[col].apply(lambda x: len(literal_eval(x)))

option_counts
```

```python
option_counts.loc[:,lang_list].mean(axis=1)
```

# Manthan

```python
lang_map = {'en':"English", 'hi':"हिंदी", 'gu':"ગુજરાતી", 'mr':"मराठी", 'bn':"বাংলা", 'te':"తెలుగు", 'ta':"தமிழ்", 'ka':"ಕನ್ನಡ"}
```

```python
manthan = questions.copy()
manthan.set_index('tag', inplace=True)
options.set_index('tag', inplace=True)
```

```python
lang_list = list(manthan.columns.difference(['tags','type']))
for tag, row in manthan.iterrows():
    for lang in lang_list:
        base_str = row['type'] + "<br 1500>" + row[lang]
        for sub_options in literal_eval(options.loc[tag,lang]):
            base_str += "<ar>"+ sub_options
        manthan.loc[tag,lang] = base_str
```

```python
manthan['options'] = options['English']
```

## Advice

```python
advice = pd.read_csv("final_advice.csv")
advice
```

```python
advice.applymap(lambda x: "\n" in x).sum(axis=1)
```

```python
advice = advice.applymap(lambda x: x.replace("\n","<br 1000>"))
```

```python
manthan_advice = advice.copy()
next_dict = {'English':'Next', 'हिंदी':'अगला', 'ગુજરાતી':'આગળ', 'বাংলা':'পরবর্তী', 'தமிழ்':'அடுத்தது', 'मराठी':'पुढे','తెలుగు':'తరువాత', 'ಕನ್ನಡ':'ಮುಂದಿನದು'}
for lang in lang_list:
    manthan_advice.loc[0,lang] = "single<br 1500>" + manthan_advice.loc[0,lang] + "<ar>" + next_dict[lang] + " - " + manthan_advice.loc[1,lang]
```

```python
manthan_advice.drop(1,axis=0, inplace=True)
```

```python
for lang in lang_list:
    manthan_advice[lang][1:] = manthan_advice[lang][1:].apply(lambda x: "single<br 1500>"+x+"<ar>"+next_dict[lang])
```
